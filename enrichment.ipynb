```python
# Dependencies
import pandas as pd
import nfl_data_py as nfl
```

```python
data = pd.read_csv('data/dataset.csv')
```

```python
drafts = nfl.import_draft_picks()
```

```python
drafts = drafts[drafts['season'] >= 2014]
```

```python
drafts = drafts[['season', 'round', 'pick', 'team', 'pfr_player_name']]

drafts = drafts.rename(columns={'pfr_player_name': 'Player', 'season': 'Draft Class'})
```

```python
# Join the dataframes
data = data.merge(drafts, on=['Player', 'Draft Class'], how='left')
```
